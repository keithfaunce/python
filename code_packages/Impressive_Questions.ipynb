### Q1. Assume that there exists a string called my_str. Write a program to check if there are duplicate characters (i.e., if any character appears more than once) in my_str. Print out the characters that are duplicate, along with their respective duplicate counts. In your output, make sure that each duplicate character (if any) is printed out only a single time.

```python
import string
alphabet_str = string.ascii_lowercase
#print(alphabet_str)
i = 0
my_str = str(input("Please enter a sentence."))
lower_sent = my_str.lower()

for i in range(len(alphabet_str)):
    letter = alphabet_str[i]
    num = lower_sent.count(letter)
    if num > 1:
        print(f"{letter} has {num} duplicates")

```

### Q7. Suppose that you are the manager of a manufacturing firm who is trying to determine the optimal price of a product based on demand. The relationship between price and demand is as follows: 
Demand = 800 – 10 x Price
#### Each product costs 50 USD to manufacture. The profit of the company is calculated as: 
Profit = (Price – Cost per item) x Demand

Write a function that, given a single price input, will first calculate its demand and then return the profit amount. Then, search a range of prices (using the function above) to identify the optimal price that will provide you the maximum profit. Notice that the range is limited, because the demand goes negative if price exceeds a certain threshold. Print out this price and the maximum profit amount.


```python
def information(price):
    per_item = 50
    demand = 800 - 10 * price
    profit = (price - per_item) * demand
    return profit

def main():
    max_profit = 0
    max_price = 0
    for price in range(0, 81):
        profit = information(price)
        if profit > max_profit:
            max_profit = profit
            best_price = price

    print(f"The best price is {best_price} with the most profit of {max_profit}")

if __name__ == "__main__":
    main()
```

### Q8. Write a program that prompts the user for a number that is more than two digits (e.g., 123, 1234, 12345, etc.). The program should divide this input into three parts (according to the rules below) and then print them as a single string joined by hyphens (-). 

- If the number of digits is divisible by 3, then each part should have equal number of digits. For instance, if the input is "123", then the output should be: “1-2-3”. 
- If the number of digits is not divisible by 3, then the first two parts should have equal number of digits whereas the last part should be longer than the first two. For instance, if the input is "54321", then the result should be "5-4-321". If the input is "1234567", then the output should be "12-34-567".

```python
num = (input("Enter a number longer than two digits"))


if len(num) % 3 == 0:
    space = len(num) // 3
    duo = space * 2
    p1 = num[:space]
    p2 = num[space:duo]
    p3 = num[duo:]
    print(p1 + "-" + p2 + "-" + p3)
else:
    space = len(num) // 3
    duo = space * 2
    p1 = num[:space]
    p2 = num[space:duo]
    p3 = num[duo:]
    print(p1 + "-" + p2 + "-" + p3)
```
