# Shape data

### Shape data - using pivot (wide data)

### Pandas pivot() function:

#### Parameters:

-  Index: Column to use to make new DataFrame's index. If None, uses exisiting index
-  Columns: Column to use to make new DataFrame's columns
-  Values: Column(s) to use for populating new DataFrame's values. If not specified all remaining columns used
-  Returns: Reshaped DataFrame
-  Exception: ValueError raised if there are any duplicate indexes

```python
# Pivot the following DataFrame: index=Year, cols=AgeGroup
# Pivoting uses ech unique value in AgeGroup as a sub-column under the other columns excluding Year the new index and the used column being AgeGroup
mortality_wide = mortality_data.pivot(index="Year",columns="AgeGroup")
mortality_wide.head(3)
```

```python
mortality_wide = mortality_data.pivot(index="Year", columns="AgeGroup", values="DeathRate")
mortality_wide.head(3)
```

```python
# Save wide DF to excel
mortality_wide.to_excel('mortality_wide.xlsx')
```

```python
# Even though Year is being used as an Index here, sending the file back to excel "re-indexes"
mortality_wide = pd.read_excel('mortality_wide.xlsx')
print(mortality_wide.head(4))
```

```python
# Save wide DF to pickle file
mortality_wide.to_pickle('mortality_wide.pkl')
```

```python
# Assign mortality_wide with new pickle
mortality_wide = pd.read_pickle('mortality_wide.pkl')
```

### Shape data - using melt (long data)

### Pandas melt() function:

#### Parameters:
  - id_Vars: Columns to use as identifier variables
  - value_vars: Columns to unpivot, if not specified uses all cols not set as id_vars
  - var_name: Name to use for variable column. If none uses frame.columns.name or variable
  - value_name: Name to use for 'value' column
  - col_level: If columns multiIndex then use this level to melt
  - ignore_index: If true, orignal index ignored. If false, original index retained
  - Returns: Reshaped DataFrame

```python
# Melt the DF mortality_wide, melting realigns AgeGroup values into the AgeGroup column
mortality_long = mortality_wide.melt(
    id_vars='Year',
    value_vars=['01-04 Years', '05-09 Years'],
    var_name='AgeGroup',
    value_name='DeathRate')

print(mortality_long.sort_values(['Year']).head(5))
```

```python
# Using pandas option_context() func, display first 3 and last 3 rows
with pd.option_context(
    'display.max_rows', 6,
        'display.max_columns', None):
    print(mortality_long)
```

# Analyze grouped data using aggregate functions

```python
# Show DeathRate comparisons between AgeGroups
mortality_data.groupby('AgeGroup').mean()
```

```python
# Show DeathRate comparisons between AgeGroups
mortality_data.groupby('Year').median(numeric_only=True).head(4)
```

```python
# Group by AgeGroup to show summary of DeathRate through describe()
mortality_data.groupby('AgeGroup')['DeathRate'].describe()
```

### Analyze grouped data using (multiple) aggregate functions

```python
# Using the agg() function, aggregate DeathRate values after grouping by AgeGroup
mortality_data.groupby('AgeGroup')['DeathRate'].agg(['mean', 'median', 'std', 'nunique'])
```

```python
# Using the agg() function, aggregate DeathRate values after grouping by Year
mortality_data.groupby('Year')['DeathRate'].agg(['mean', 'median', 'std', 'min', 'max', 'var', 'nunique'])
```

```python
# Formatting
mortality_wide = mortality_wide.set_index('Year')

mortality_wide.to_pickle('mortality_wide.pkl')
mortality_wide = pd.read_pickle('mortality_wide.pkl')
mortality_wide.head()
```

```python
# Create a box plot to show IQR, puts emphasis on outliers
mortality_wide.plot.box()
```

```python
# DeathRate by AgeGroup visualized per year
mortality_wide.plot(xlabel='Year', ylabel='Deaths per 100,000', title='DeathRate by AgeGroup')
```

```python
# Bar chart shows dramatic differences between mean, median and std. Showcasing how effective medical practices have become over ~100 years
mortality_data.groupby('AgeGroup')['DeathRate'].agg(['mean', 'median', 'std']).plot.bar(rot=45)
```

```python
# Same chart, but horizontal
mortality_data.groupby('AgeGroup')['DeathRate'].agg(['mean', 'median', 'std']).plot.barh()
```

```python
# Showcase comparisons on how medical practice has evoloved over instances of 25 years
mortality_data.query('Year in (1900, 1925, 1950, 1975, 2000)').groupby('Year').DeathRate.sum().plot.pie()
```

```python
# Child Mortality values grouped by AgeGroup over 25 year increments
# Year is the y-axis, Number of Deaths (per 100,000) is the x-axis
mortality_wide.query('Year in (1900, 1925, 1950, 1975, 2000)').plot.barh(
    title=['Child Mortality: 01-04', 'Child Mortality: 05-09',
           'Child Mortality: 10-14', 'Child Mortality: 15-19'],
    sharey=True, legend=False, subplots=True, layout=(2, 2), figsize=(8,5))
```

```python
# Using a dot plot to highlight a positive trend in medical practices for infants
mortality_data.query('AgeGroup == "01-04 Years"').plot.scatter(x='Year', y='DeathRate', c='Green')
```
