### Question: 
You are asked to build a simple inventory management system for storing and updating information about a retailer's products. Assume that this retail offers 5 products. For each product, the system stores a record of product inventory level (quantity) and the price. This information is stored in a file called "inventory.txt" (attached to the assignment). <br>

In this file, each line represents a unique product following the format of: Product Name, Quantity, Price. Product characteristics are separated with comma (,) delimiters. <br>

There are three main operations provided by the system (listed below). When the system is started, the program should list these three operations and let the users choose which one to execute. These functionalities are as follows:<br>

1) Generate overall statistics for the <u>entire inventory:</u> total quantity of all items, average price per item, the product with the minimum inventory (and its quantity) and the product with the maximum inventory (and its quantity).

2) Retrieve information (quantity, price, or both) for a given product

3) Update information (quantity or price) for a given product

Once the user makes a selection, the program should invoke a relevant <u><b>function</b></u> and complete the operation. The system should continue running (i.e., functionality selection should be provided to the user repeatedly) until the user kills the execution of the program (possibly with a unique input character). A suggested system design will have four functions:

<b>Function 1</b>: display_system_operations<br>
Inputs: None<br>
Outputs: Print out operation information<br>
Description: This function prints out all possible operations of the system for the user to choose.

<b>Function 2</b>: generate_overall_stats<br>
Inputs: None<br>
Outputs: Print overall statistics<br>
Description: This function generates the overall statistics for the <u>entire inventory</u>: total quantity of all items, average price per item (across the entire inventory), the product with the minimum inventory (and its quantity) and the product with the maximum inventory (and its quantity).

<b>Function 3</b>: retrieve_information<br>
Inputs: product name and information type (quantity or price or both)<br>
Outputs: Print the information requested<br>
Description: This function retrieves the relevant information for a given product, which is obtained from the user as an input.

<b>Function 4</b>: update_information<br>
Inputs: product name, information type (quantity or price), and the updated value<br>
Outputs: Write to inventory file<br>
Description: This function updates the quantity or the price of a given product in the inventory file.

<b>Function 5</b>: main<br>
Inputs: None<br>
Outputs: Program execution<br>
Description: The main function first imports (i.e., reads) the inventory text file into the memory and asks the user which of the specific operations / functionality that the user would like to perform. It should invoke the relevant function (described above) for the operation chosen by the user. Finally, it should ask the user if they want to re-run the program again or not. If not, then terminate.


```python
def display_system_operations(inv_dict, choice):
    #view_inv = str(input("To view entire inventory: total product"))
    #if view_inv.lower() == 'total product':
    if choice.lower() == 'inventory':
        generate_overall_stats(inv_dict)
    elif choice.lower() == 'product':
        retrieve_information(inv_dict)
    elif choice.lower() == 'update':
        update_information(inv_dict)

def generate_overall_stats(inv_dict):
    total_product = 0
    total_price = 0
    max_val = 0
    min_val = 1000
    for item in inv_dict.values():
        total_product += item['amount']
    for item in inv_dict.values():
        item_total = item['amount'] * item['price']
        total_price += item_total
    avg_price = round(float(total_price) / total_product, 2)
    print(f"The total number of products is {total_product}. The average price is ${avg_price}.")
    
    for key, item in inv_dict.items():
        if item['amount'] > max_val:
            max_val = item['amount']
            max_key = key
    for keys, value in inv_dict.items():
        if item['amount'] < min_val:
            min_val = value['amount']
            min_key = keys
    print(f"We have the most {max_key} with {max_val}. We have the least {min_key} with {min_val}.")
    

def retrieve_information(inv_dict):
    for key2 in inv_dict.keys(): # to show products avaliable
        print(key2)
    option = str(input("Please select a product by entering its name.")).lower()
    amtpr = str(input("To view how much of a product we have type: a To view how much a product costs type: c For both type: b")).lower()
    for key2 in inv_dict.keys():
        if option == key2.lower():
            if amtpr == 'a':
                print(f"We have {inv_dict[key2]['amount']} of {key2}")
            elif amtpr == 'c':
                print(f"The {key2} are {inv_dict[key2]['price']} apiece.")
            elif amtpr == 'b':
                print(f"We have {inv_dict[key2]['amount']} {key2}, they're {inv_dict[key2]['price']} apiece.")

    print("If there were no results, there is a spelling mistake in your inputs.")


def update_information(inv_dict):
    for key2 in inv_dict.keys(): # to show products avaliable
        print(key2)
    option = str(input("Please select a product by entering its name.")).lower()
    amtpr = str(input("To change product quantity type: a To change how much a product costs type: c To change both type: b")).lower()
    for key2 in inv_dict.keys():
        if option == key2.lower():
            if amtpr == 'a':
                new_amt = int(input(f"We currently have {inv_dict[key2]['amount']} of {key2}. Please update the amount."))
                inv_dict[key2]['amount'] = new_amt
            elif amtpr == 'c':
                new_price = float(input(f"The current price of {key2} is {inv_dict[key2]['price']}. Please update the price."))
                inv_dict[key2]['price'] = new_price
            elif amtpr == 'b':
                new_amt = int(input(f"We currently have {inv_dict[key2]['amount']} of {key2}. Please update the amount."))
                new_price = float(input(f"The current price of {key2} is {inv_dict[key2]['price']}. Please update the price."))
                inv_dict[key2]['price'] = new_price
                inv_dict[key2]['amount'] = new_amt

    inventory = open('C:/Users/Keith/MS_BA-Sem1/ism_sum/txts/inventory.txt', 'w')
    for key2, value in inv_dict.items():
        newlines = f"{key2},{value['amount']},{value['price']}\n"
        inventory.write(newlines)
    inventory.close()

def main():
    inv_dict = {}
    inventory = open('C:/Users/Keith/MS_BA-Sem1/ism_sum/txts/inventory.txt', 'r')
    for line in inventory:
        line = line.strip() #removes \n that appears when pulling values bc it counts as whitespace when no args
        if line:
            name, amount, price = line.split(',')
            #print(amount)
            inv_dict[name] = {'amount': int(amount), 'price': float(price)}
    inventory.close()
    #print(type(inv_dict))
    #print(inv_dict)
    #return inv_dict
    #choice = str(input("Options, type to execute. 1.Check on total inventory: inventory  2.See product details: product"))
    #flagStaff = False
    while True:
        choice = str(input("Options, type to execute. Check on total inventory: inventory  See product details: product  To update product info: update"))
        if choice.lower() == 'inventory':
            #flagStaff = True
            display_system_operations(inv_dict, choice)
        elif choice.lower() == 'product':
            #flagStaff = True
            display_system_operations(inv_dict, choice)
        elif choice.lower() == 'update':
            #flagStaff = True
            display_system_operations(inv_dict, choice)
        else:
            #flagStaff = False
            print("Please select a correct option")

        repeat = str(input("To perform another operation type: y If not, type: n")).lower()
        if repeat != 'y':
            break

if __name__ == "__main__":
    main()
```
